# Ising RG transformation plots

We plot the Renormalization Group (RG) transformation $K'(K)$ for
* the standard 1-dim Ising model
* the standard 2-dim Ising model
* the 2-dim Ising-dbnet (dbnet= Dynamic Bayesian Network)

The 1-dim model has no fixed point other than $K=0$.

Both of the 2-dim models have  a repulsive fixed point (i.e., critical point) near although not exacty 
at the Curie temperature coupling constant $\beta_c J=K_c\approx 0.44$
