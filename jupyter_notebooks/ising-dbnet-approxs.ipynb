# ising-dbnet: approximations of the RG transformation.

For the Ising-dbnet model, we used Sympy to find the
Taylor series and asymptotic limit of the RG transformation

```python
from sympy import *

Pp, lam, eta, K = symbols('Pp lam eta K')
Pm = 1 - Pp

Denom = Pp/cosh(K*(1+lam))**4 + \
        Pm/cosh(K*(-1+lam))**4
Num = Pp*exp(K*eta*(1+4*lam))/cosh(K*(1+lam))**4 + \
      Pm*exp(K*eta*(-1+4*lam))/cosh(K*(-1+lam))**4

Kprime = (log(Num) - log(Denom))/(4*eta*lam)
```

```python
Kprime_s = series(Kprime, K, 0, 3)
```

```python
print(Kprime_s)
```

```python
simplify(Kprime_s)
```

```python
Kprime = Kprime.subs({lam:Rational(1, 8), eta:Integer(4)})

Kprime
```

```python
Kprime_exp =Kprime.rewrite(exp)
Kprime_asym = series(Kprime_exp, K, oo, 2)
Kprime_asym
```

```python
logcombine(Kprime_asym)
```
