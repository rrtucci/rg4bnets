# ising-dbnet: Taylor series and limit of  RG transformation

```python
from sympy import *

Pp, lam, eta, K = symbols('Pp lam eta K')
Pm = 1 - Pp

Denom = Pp/cosh(K*(1+lam))**4 + \
        Pm/cosh(K*(-1+lam))**4
Num = Pp*exp(K*eta*(1+4*lam))/cosh(K*(1+lam))**4 + \
      Pm*exp(K*eta*(-1+4*lam))/cosh(K*(-1+lam))**4

Kprime = (log(Num) - log(Denom))/(4*eta*lam)
```

```python
Kprime_s = series(Kprime, K, 0, 3)
```

```python
print(Kprime_s)
```

```python
simplify(Kprime_s)
```
